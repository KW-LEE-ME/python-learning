### Step 1. Import the necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
path = '../../../../pandas_exercises_with_data/data/'
titanic = pd.read_csv(path + 'titanic.csv')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId', inplace=True)
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic['Sex'].value_counts(), labels=['male', 'female'], autopct='%1.2f%%')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.lmplot('Age', 'Fare', hue='Sex', data=titanic, fit_reg=False)
plt.show()
```

```python
titanic['Sex_'] = (titanic['Sex'] == 'male').astype('int')
```

```python
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Sex_'], alpha=.7)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
