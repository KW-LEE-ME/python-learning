### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
path = '../../../../pandas_exercises_with_data/data/'
chipo = pd.read_csv(path + "chipotle.tsv", sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top5 = chipo.groupby('item_name').sum().sort_values(by='quantity', ascending=False)[:5]
plt.figure(figsize=(10, 8))
top5['quantity'].plot(kind='bar')
plt.xlabel('item_name')
plt.ylabel('quantity')
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x:float(x[1:]))
grouped = chipo.groupby('order_id').sum()
plt.figure(figsize=(10, 8))
plt.scatter(grouped['item_price'], grouped['quantity'])
plt.xlabel('price')
plt.ylabel('quantity')
plt.show()
```
